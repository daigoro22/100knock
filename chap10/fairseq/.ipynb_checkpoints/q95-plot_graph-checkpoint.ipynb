```python
import numpy as np
import matplotlib.pyplot as plt

if __name__ == "__main__":
    x_range=0
    score_list=[]
    with open("q95-bpe-bleu_log.txt") as f:
        for line in f:
            line=line.split()
            if(line[0]=="BLEU4"):
                bleu_score=float(line[2].rstrip(","))
                x_range+=1
                score_list.append(float(bleu_score))
    x_axis=[x for x in range(1,x_range+1)]
    print(score_list)
    #left = np.array([1, 2, 3, 4, 5])
    #height = np.array([100, 300, 200, 500, 400])
    #plt.plot(left, height)
    
    
    plt.plot(x_axis, score_list)
    plt.show()
```
